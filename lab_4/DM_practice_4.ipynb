#### **Система Лоренца:**

Система Лоренца описується трьома диференціальними рівняннями:

$$
\begin{cases}
\frac{dx}{dt} = \sigma (y - x) \\
\frac{dy}{dt} = x (\rho - z) - y \\
\frac{dz}{dt} = xy - \beta z
\end{cases}
$$

де:
- $( \sigma )$ — коефіцієнт Прандтля (зазвичай $ \sigma = 10 $),
- $( \rho )$ — числовий параметр для турбулентності (зазвичай $ \rho = 28 $),
- $( \beta )$ — коефіцієнт відношення висоти до довжини (зазвичай $ \beta = 8/3 $).

#### **Об'єкти Нарендра-Пархтізаратхі:**

Об'єкти Нарендра-Пархтізаратхі описуються наступними рівняннями:

$$
\begin{cases}
x_1(t+1) = \alpha_1 x_1(t) + \beta_1 x_2(t) + \gamma_1 x_3(t) + \delta_1 \\
x_2(t+1) = \alpha_2 x_1(t) + \beta_2 x_2(t) + \gamma_2 x_3(t) + \delta_2 \\
x_3(t+1) = \alpha_3 x_1(t) + \beta_3 x_2(t) + \gamma_3 x_3(t) + \delta_3
\end{cases}
$$

де $(\alpha_i)$, $(\beta_i)$, $(\gamma_i)$ і $(\delta_i)$ — коефіцієнти для кожного з рівнянь.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import Ridge
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.multioutput import MultiOutputRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#1.1

sigma = 10.0
rho = 28.0
beta = 8.0 / 3.0

def lorenz(state, t):
    x, y, z = state
    dx = sigma * (y - x)
    dy = x * (rho - z) - y
    dz = x * y - beta * z
    return [dx, dy, dz]

initial_state = [1.0, 1.0, 1.0]
t = np.linspace(0, 50, 10000) 

lorenz_data = odeint(lorenz, initial_state, t)

x, y, z = lorenz_data[:, 0], lorenz_data[:, 1], lorenz_data[:, 2]

fig = plt.figure(figsize=(8, 6))
ax = fig.add_subplot(projection='3d')
ax.plot(x, y, z, lw=0.5)
ax.set_title('Lorenz Attractor')
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
plt.show()
```

```python
#1.2

a1 = 1.0
a2 = 0.5
b1 = 0.5
b2 = 2.0

def np_system(state, t, a1, a2, b1, b2):
    x1, x2 = state
    dx1_dt = -a1 * x1 + b1 * x2**3
    dx2_dt = a2 * np.sin(x1) - b2 * x2
    return [dx1_dt, dx2_dt]

initial_state_np = [1.0, 1.0]

t_np = np.linspace(0, 50, 10000)

np_states = odeint(np_system, initial_state_np, t_np, args=(a1, a2, b1, b2))

x1 = np_states[:, 0]
x2 = np_states[:, 1]

plt.figure(figsize=(10, 4))
plt.plot(t_np, x1, label='$x_1$')
plt.plot(t_np, x2, label='$x_2$')
plt.title('Narendra-Parthasarathy System (V3)')
plt.xlabel('Time')
plt.ylabel('State Variables')
plt.legend()
plt.show()
```

```python
#2.1

def create_supervised_data(data, input_steps=3, output_steps=1):
    X, y = [], []
    for i in range(len(data) - input_steps - output_steps + 1):
        X.append(data[i:i + input_steps].flatten())
        y.append(data[i + input_steps:i + input_steps + output_steps].flatten())
    return np.array(X), np.array(y)

lorenz_data_combined = np.column_stack((x, y, z))
X_lorenz, y_lorenz = create_supervised_data(lorenz_data_combined)

np_data_combined = np.column_stack((x1, x2))
X_np, y_np = create_supervised_data(np_data_combined)

def split_data(X, y):
    X_train_val, X_test, y_train_val, y_test = train_test_split(
        X, y, test_size=0.15, random_state=42)
    X_train, X_val, y_train, y_val = train_test_split(
        X_train_val, y_train_val, test_size=0.1765, random_state=42)
    return X_train, X_val, X_test, y_train, y_val, y_test

X_train_lorenz, X_val_lorenz, X_test_lorenz, y_train_lorenz, y_val_lorenz, y_test_lorenz = split_data(X_lorenz, y_lorenz)
X_train_np, X_val_np, X_test_np, y_train_np, y_val_np, y_test_np = split_data(X_np, y_np)

print("Lorenz System Data Shapes:")
print(f"X_train: {X_train_lorenz.shape}, y_train: {y_train_lorenz.shape}")
print(f"X_val: {X_val_lorenz.shape}, y_val: {y_val_lorenz.shape}")
print(f"X_test: {X_test_lorenz.shape}, y_test: {y_test_lorenz.shape}")
```

```python
#2.2

def normalize_data(X_train, X_val, X_test, y_train, y_val, y_test):
    scaler_X = StandardScaler().fit(X_train)
    scaler_y = StandardScaler().fit(y_train)
    X_train_scaled = scaler_X.transform(X_train)
    X_val_scaled = scaler_X.transform(X_val)
    X_test_scaled = scaler_X.transform(X_test)
    y_train_scaled = scaler_y.transform(y_train)
    y_val_scaled = scaler_y.transform(y_val)
    y_test_scaled = scaler_y.transform(y_test)
    return X_train_scaled, X_val_scaled, X_test_scaled, y_train_scaled, y_val_scaled, y_test_scaled, scaler_X, scaler_y

X_train_lorenz_scaled, X_val_lorenz_scaled, X_test_lorenz_scaled, \
y_train_lorenz_scaled, y_val_lorenz_scaled, y_test_lorenz_scaled, \
scaler_X_lorenz, scaler_y_lorenz = normalize_data(
    X_train_lorenz, X_val_lorenz, X_test_lorenz, y_train_lorenz, y_val_lorenz, y_test_lorenz)

print("Mean of X_train_lorenz_scaled:", np.mean(X_train_lorenz_scaled))
print("Std of X_train_lorenz_scaled:", np.std(X_train_lorenz_scaled))
```

```python
#3.1

models = {
    'Ridge Regression': Ridge(),
    'Decision Tree': DecisionTreeRegressor(),
    'Random Forest': RandomForestRegressor(),
    'SVM': MultiOutputRegressor(SVR())
}
```

```python
#3.2

def train_models(models, X_train, y_train):
    trained_models = {}
    for name, model in models.items():
        model.fit(X_train, y_train)
        trained_models[name] = model
    return trained_models

trained_models_lorenz = train_models(models, X_train_lorenz_scaled, y_train_lorenz_scaled)
```

```python
#4.1

param_grids = {
    'Ridge Regression': {
        'alpha': [0.1, 1.0, 10.0],
        'solver': ['auto', 'svd', 'cholesky']
    },
    'Decision Tree': {
        'max_depth': [None, 5, 10],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2]
    },
    'Random Forest': {
        'n_estimators': [50, 100],
        'max_depth': [None, 10],
        'min_samples_split': [2, 5],
        'min_samples_leaf': [1, 2]
    },
    'SVM': {
        'estimator__C': [0.1, 1],
        'estimator__kernel': ['linear', 'rbf'],
        'estimator__gamma': ['scale', 'auto']
    }
}

def grid_search_models(models, param_grids, X_train, y_train):
    best_models = {}
    for name, model in models.items():
        grid = GridSearchCV(model, param_grids[name], cv=3, scoring='neg_mean_squared_error', n_jobs=-1)
        grid.fit(X_train, y_train)
        best_models[name] = grid.best_estimator_
        print(f"{name} Best Parameters: {grid.best_params_}")
    return best_models

best_models_lorenz = grid_search_models(models, param_grids, X_train_lorenz_scaled, y_train_lorenz_scaled)
```

```python
#5.1

def evaluate_models(models, X_val, y_val):
    metrics = {}
    for name, model in models.items():
        y_pred = model.predict(X_val)
        mse = mean_squared_error(y_val, y_pred)
        r2 = r2_score(y_val, y_pred)
        metrics[name] = {'MSE': mse, 'R2': r2}
        print(f"{name} - Validation MSE: {mse:.4f}, R2 Score: {r2:.4f}")
    return metrics

validation_metrics = evaluate_models(best_models_lorenz, X_val_lorenz_scaled, y_val_lorenz_scaled)
```

```python
#5.2

test_metrics = evaluate_models(best_models_lorenz, X_test_lorenz_scaled, y_test_lorenz_scaled)
```

```python
#6.1

results_dataframe = pd.DataFrame({
    'Model': list(validation_metrics.keys()),
    'Validation MSE': [m['MSE'] for m in validation_metrics.values()],
    'Validation R2': [m['R2'] for m in validation_metrics.values()],
    'Test MSE': [m['MSE'] for m in test_metrics.values()],
    'Test R2': [m['R2'] for m in test_metrics.values()]
})

print(results_dataframe)
```

```python
#6.2

model_names = ['Ridge Regression', 'Decision Tree', 'Random Forest', 'SVM']

y_test_true = scaler_y_lorenz.inverse_transform(y_test_lorenz_scaled)

for name in model_names:
    model = best_models_lorenz[name]
    y_test_pred_scaled = model.predict(X_test_lorenz_scaled)
    y_test_pred = scaler_y_lorenz.inverse_transform(y_test_pred_scaled)
    
    plt.figure(figsize=(12, 6))
    plt.plot(y_test_true[:100, 0], label='Actual', color='blue')
    plt.plot(y_test_pred[:100, 0], label='Predicted', linestyle='--', color='red')
    plt.title(f'Actual vs Predicted Values ({name} - Lorenz System X Component)')
    plt.xlabel('Sample Index')
    plt.ylabel('X Value')
    plt.legend()
    plt.show()
```
